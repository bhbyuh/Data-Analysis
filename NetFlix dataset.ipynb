```python
#2-

integer_columns = NF.select_dtypes(include=['object']).columns.tolist()
print("Column with Numeric Data Type are \n",integer_columns)
integer_columns = NF.select_dtypes(include=['int']).columns.tolist()
print("\nColumn with Non-Numeric Data Type are integer_columns \n",integer_columns)
```

```python
#3-
columnslist=NF.columns
DFlength=len(NF)
columnlenindex=0
for i in columnslist:
    total_count_per_col=NF[i].count()
    print(i," column have ",DFlength-total_count_per_col," NULL values")
```

```python
#4
showcount=0
moviescount=0
for i in NF['type']:
    if(i=='Movie'):
        moviescount+=1
    if(i=='TV Show'):
        showcount+=1
        
print("No of Movies ",moviescount)
print("No of TV Shows ",showcount)
```

```python
#5
count=len(NF)-1
Indiacount=0
UScount=0
while(count>=0):
    if(NF.loc[count,'country']=='United States' and NF.loc[count,'type']=='Movie'):
        UScount+=1
    if(NF.loc[count,'country']=='India' and NF.loc[count,'type']=='Movie'):
        Indiacount+=1
    count-=1
        
print("No of Movies In India",Indiacount)
print("No of Movies In United States",UScount)
```

```python
#6
import numpy as np
directr=[]
directors=NF['director'].unique()
for i in directors:
    if(i!=np.nan):
         directr.append(i)
    
print(len(directr))
directr
```

```python
#7
count=len(NF)-1
MVcount2019=0
MVcount2020=0
MVcount2021=0
while(count>=0):
    if(NF.loc[count,'release_year']==2019 and NF.loc[count,'type']=='Movie'):
        MVcount2019+=1
    if(NF.loc[count,'release_year']==2020 and NF.loc[count,'type']=='Movie'):
        MVcount2020+=1
    if(NF.loc[count,'release_year']==2021 and NF.loc[count,'type']=='Movie'):
        MVcount2021+=1
    count-=1
        
print("No of Movies release in 2019",MVcount2019)
print("No of Movies release in 2020",MVcount2020)
print("No of Movies release in 2021",MVcount2021)
```

```python
#8
totallength=len(NF)-1
Movie_count=0
Sub_Movie_count=0
movie_index=0
for i in NF['type']:
    j=0
    movie=NF.loc[movie_index,'title']
    Sub_Movie_count=0
    if(i=='Movie'):
        while(j<=totallength):
            if(NF.loc[j,'release_year']>=2010 and NF.loc[j,'release_year']<=2021 and NF.loc[j,'title']==movie):
                Sub_Movie_count+=1
            j+=1
        if(Sub_Movie_count>=12):
            Movie_count+=1
    movie_index+=1
print("Number of movies relies every year Between 2010 to 2021 are ",Movie_count)
```

```python
#9
import datetime
date1=datetime.date(2015,7,1)
date2=datetime.date(2021,7,1)
NO_of_Movies=0
added_date_index=0
import numpy as np
day=[]
month=[]
year=[]
date_added=[]
for i in NF['date_added']:
    if(i!=np.nan):
        j=str(i)
        (Month,Day,Year)=j.split(" ")
        Day=Day[:-1]
        day.append(int(Day))
        year.append(int(Year))
        month.append(Month)
    else:
        day.append("")
        year.append("")
        month.append("")
NF['month']=month
NF['YEAR']=year
NF['day']=day
NF['month']=NF['month'].replace({"January":1,"February": 2,"March": 3,"April": 4,"May": 5,"June":6 ,"July": 7,
        "August": 8,
        "September": 9,
        "October": 10,
        "November": 11,
        "December": 12})
i=0
totallength=len(NF)-1
while(i<=totallength):
    if(NF.loc[i,'YEAR']!=np.nan):
        date_added.append(datetime.date(NF.loc[i,'YEAR'], NF.loc[i,'month'], NF.loc[i,'day']))
        i+=1
    else:
         date_added.append("")
            
NF['date_added']=date_added

for i in NF['type']:
    if(i=='Movie'):
        if(NF.loc[added_date_index,'date_added']>=date1 and NF.loc[added_date_index,'date_added']<=date2):
            NO_of_Movies+=1
    added_date_index+=1
    

print("No of Movies Bwtween 1 July,2015 and 1 July,2021 are ",NO_of_Movies)
NF=NF.drop('month',axis=1)
NF=NF.drop('YEAR',axis=1)
NF=NF.drop('day',axis=1)
```
